```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import cv2
import numpy as np
from src.video_loader import load_video
from src.detection import detect_objects
from src.pass_detector import detect_pass

video_path = "../data/match.mp4"

previous_ball = None
frames_processed = 0

for i, frame in enumerate(load_video(video_path)):

    # Stop after 80 frames
    if i > 80:
        break

    # Process every 5th frame
    if i % 5 != 0:
        continue

    players, ball = detect_objects(frame)

    pass_event = False

    if ball is not None and previous_ball is not None:

        speed = np.linalg.norm(
            np.array(ball[:2]) - np.array(previous_ball[:2])
        )

        if speed > 25:
            pass_event = True
            print(f"⚽ Pass detected at frame {i}")

    previous_ball = ball
    frames_processed += 1

print("Frames processed:", frames_processed)
```
