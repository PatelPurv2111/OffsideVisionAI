```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import cv2
from src.detection import detect_objects
from src.visualization import draw_players, draw_ball
from matplotlib import pyplot as plt

video_path = "../data/match.mp4"

cap = cv2.VideoCapture(video_path)

ret, frame = cap.read()

players, ball = detect_objects(frame)

draw_players(frame, players)
draw_ball(frame, ball)

frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(frame_rgb)
plt.axis("off")

print("Players detected:", len(players))
print("Ball detected:", ball)
```
